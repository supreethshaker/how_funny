```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
df = df.drop(columns='Unnamed: 0')
df
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df,
                x=df['Train F1'],
                y=df['Test F1'],
                hue='Model Name',
                style='Preprocessing',
                palette='colorblind',
                s=200
                )
```

```python
labels = df['Preprocessing'].unique()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

axes[0].set_title("F1 Scores by Model and Dataset")
sns.scatterplot(data=df,
                x=df['Train F1'],
                y=df['Test F1'],
                hue='Model Name',
                style='Preprocessing',
                palette='colorblind',
                s=200,
                ax=axes[0]
                )


axes[1].set_title("F1-Scores for Top 5 Models")
tmp_df = df.sort_values(by=['Test F1']).tail(5)
sns.scatterplot(data=tmp_df,
                x=tmp_df['Train F1'],
                y=tmp_df['Test F1'],
                hue='Model Name',
                style='Preprocessing',
                palette='colorblind',
                s=200,
                ax=axes[1]
                )
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,
                x=df['Train F1'],
                y=df['Test F1'],
                hue='Model Name',
                style='Preprocessing',
                palette='colorblind',
                s=200
                )
plt.title("Test F1 Score for all Models")
```

```python
labels = df['Model Name'].unique()
```

```python
for i in range(len(labels)):
    tmp_df = df[df['Model Name'] == labels[i]]
    sns.scatterplot(data=df,
                    x=tmp_df['Train F1'],
                    y=tmp_df['Test F1'],
                    hue='Preprocessing',
                    style='Preprocessing',
                    palette='colorblind',
                    s=200
                    )
    plt.title("F1-Scores for %s" %(labels[i]))
    plt.savefig("figures/%s.png" %(labels[i]))
    plt.show()
```

```python
tmp_df = df.sort_values(by=['Test F1']).tail(5)
sns.scatterplot(data=tmp_df,
                x=tmp_df['Train F1'],
                y=tmp_df['Test F1'],
                hue='Model Name',
                style='Preprocessing',
                palette='colorblind',
                s=200
                )
plt.title("F1-Scores for the Top 5 Models")
plt.savefig("figures/top-5.png")
```
